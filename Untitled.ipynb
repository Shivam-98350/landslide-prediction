## Applying Logistic Regression Using train-test-split

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
X_train,X_test,y_train,y_test = train_test_split(x_scaled,y,test_size=.20,random_state=10)
```

```python
model=LogisticRegression(C=10, penalty='elasticnet',l1_ratio=.6,solver='saga',multi_class='multinomial',max_iter=100000)
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score,make_scorer
y_pred = model.predict(X_test)
accuracy_logistic = accuracy_score(y_test, y_pred)
accuracy_df['Model'].append('Logistic Regression')
accuracy_df['Accuracy'].append(accuracy_logistic)
accuracy_df['Precision'].append(precision_score(y_test,y_pred,average='weighted'))
accuracy_df['Recall'].append(recall_score(y_test,y_pred,average='weighted'))
accuracy_df['F1_Score'].append(f1_score(y_test,y_pred,average='weighted'))
print(accuracy_logistic)
print(accuracy_df)
```

```python
df_classi=pd.DataFrame(classification_report(y_test,y_pred,zero_division=0,output_dict=True))
df_classi
```

```python
from sklearn.metrics import plot_confusion_matrix

print("Confusion Matrix : \n",
confusion_matrix(y_test,y_pred,))
print("\nAccuracy : ",
accuracy_score(y_test,y_pred)*100,"\n")
print("Report :\n",
classification_report(y_test,y_pred,zero_division=0))
plot_confusion_matrix(model,X_train,y_train)
```

## Applying Logistic Regression Using Stratified K-Fold

```python
lr = LogisticRegression(C=6, penalty='elasticnet',l1_ratio=.6,solver='saga',multi_class='multinomial',max_iter=100000)
```

```python
from sklearn.model_selection import StratifiedKFold,cross_validate
skf = StratifiedKFold(n_splits=6, shuffle=True, random_state=1)
lst_accu_stratified = []
```

```python
from statistics import mean, stdev
```

```python
for train_index, test_index in skf.split(X, y):
    x_train_fold, x_test_fold = x_scaled[train_index], x_scaled[test_index]
    y_train_fold, y_test_fold = y[train_index], y[test_index]
    lr.fit(x_train_fold, y_train_fold)
    lst_accu_stratified.append(lr.score(x_test_fold, y_test_fold))
                              
k_fold_logistic_accuracy = mean(lst_accu_stratified)
accuracy_df['Model'].append('K-Fold Logistic Regression')
accuracy_df['Accuracy'].append(k_fold_logistic_accuracy)
accuracy_df['Precision'].append(k_fold_logistic_accuracy)
accuracy_df['Recall'].append(k_fold_logistic_accuracy)
accuracy_df['F1_Score'].append(k_fold_logistic_accuracy)

# Print the output.
print('List of possible accuracy:', lst_accu_stratified)
print('\nMaximum Accuracy That can be obtained from this model is:',
      max(lst_accu_stratified)*100, '%')
print('\nMinimum Accuracy:',
      min(lst_accu_stratified)*100, '%')
print('\nOverall Accuracy:',
      mean(lst_accu_stratified)*100, '%')
print('\nStandard Deviation is:', stdev(lst_accu_stratified))
```
